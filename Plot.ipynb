# Accuracy to clean up steps

```python
# name: {model}{level of difficulty}
x = [2000 * i for i in range(6)]
sgm5 = [0, 0, .2, .2, .35, .5]
sgm1 = [.75, .85, .75, .7, .85, .45]
sgm2 = [.2, .25, .15, .45, .55, .5]
sgm3 = [.05, .15, 0, .4, .4, .6]
plt.plot(x, sgm1, color='lightgrey', label='level 1')
plt.plot(x, sgm2, color='darkgray', label='level 2')
plt.plot(x, sgm3, color='dimgrey', label='level 3')
plt.plot(x, sgm5, color='black', label='level 4')
plt.legend()
plt.xlabel('clean-up steps')
plt.ylabel('success rate')
plt.figure(figsize=(40,10))
```
